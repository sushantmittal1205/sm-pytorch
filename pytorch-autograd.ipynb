### PyTorch Autograd is used to compute gradients during the backward pass.
Autograd is a core component of PyTorch that provides automatic differentiation for tenson operations.

```python
import torch
print(torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU Exists:", torch.cuda.get_device_name())
```

```python
# Examples
#1. y = x**2
```

```python
# Requires Gradient (requires_grad) will instruct PyTorch to calculate derivative for this Tensor.

# Define x and y
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2
```

```python
# Display x and y
x, y
```

```python
# Compute Gradient (dy_dx)
y.backward()
```

```python
# Display Gradient 
x.grad
```

```python
# A little complex example
# 2. y = x**2, z = sin(y)
import math
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2
z = torch.sin(y)
```

```python
x, y, z
```

```python
# Compute Gradient (dz_dz = dz_dy * dy_dx)
z.backward()
```

```python
# Display Gradient 
x.grad
```

```python
# 3. Neural Network

x = torch.tensor(8.45)
y = torch.tensor(1.0)

print(x)
print(y)
```

```python
# Parameters
w = torch.rand(1, requires_grad=True)
b = torch.rand(1, requires_grad=True)

print(w)
print(b)
```

```python
# Loss Function
def loss_function(y_true, y_pred):
    return 0.5*(y_true-y_pred)**2
```

```python
# Forward Pass
z = w * x + b
y_pred = torch.sigmoid(z)
loss = loss_function(y, y_pred)
```

```python
print(z)
print(y_pred)
print(loss)
```

```python
# Compute Gradient
loss.backward()
```

```python
# Print Loss gradients w.r.t parameters

#d_loss_d_w
print(w.grad)

#d_loss_d_b
print(b.grad)
```

```python
# How to clear gradients (To prevent accumulating gradients.)
w.grad.zero_()
w.grad
```

```python
# Diable gradient tracking
x = torch.tensor(20.0, requires_grad=True)
x
```

```python
y = x ** 2
y
```

```python
y.backward()
print(x.grad)
```

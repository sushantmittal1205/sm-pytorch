### Creating a Tensor

```python
# Empty
torch.empty(2,3)
```

```python
# Zeros
torch.zeros(2,3)
```

```python
# Ones
torch.ones(2,3)
```

```python
# Random
torch.rand(2,3)
```

```python
# Setting Seed to get same matrix values for reproducibility.

torch.manual_seed(0)
x = torch.rand(2,3)

torch.manual_seed(0)
y = torch.rand(2,3)

x==y
```

```python
# Torch from List

torch.tensor([[1,2,3],[4,5,6]])
```

```python
# Arange
torch.arange(2,10,2)
```

```python
# Even Spacing
torch.linspace(1, 10, 10)
```

```python
# Identity Matrix
torch.eye(2,2)
```

```python
# Full
torch.full((2,3), 47)
```

### Tensor Shapes

```python
x = torch.empty(2,8)
print(x)
print(x.shape)
```

```python
x = torch.empty(2,8)
print(x)
print(x.size())
```

```python
# Create Same Shape Tensors
y = torch.empty_like(x)
print(y)
print(y.shape)
```

```python
z = torch.zeros_like(x)
print(z)
print(z.shape)
```

```python
zz = torch.ones_like(x)
print(zz)
print(zz.shape)
```

```python
zzz = torch.rand_like(x)
print(zzz)
print(zzz.shape)
```

```python
zzz = torch.rand_like(x)
print(zzz)
print(zzz.shape)
```

### Tensor Data Types

```python
x.dtype
```

```python
# Create Int type Tensor
torch.tensor([1.0, 2.0, 3.0], dtype=torch.int32)
```

```python
torch.tensor([1,2,3,4], dtype=torch.float16)
```

```python
torch.tensor([1.0, 2.0, 3.0], dtype=torch.int8)
```

```python
# Change Data Type of existing tensor
x = 12 + torch.rand(2,2)
print(x.dtype)
print(x)

y = x.to(torch.int8)
print(y.dtype)
print(y)
```

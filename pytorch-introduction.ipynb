### Mathematical Operations

```python
x = torch.rand(2,3)
print(x)
```

#### 1. Scalar Operations

```python
print(x)

# addition
print(x+2)

# substraction
print(x-2)

# multiplication
print(x*10)

# division
print(x/0.10)

# int division
print(x*100//3)

# mod
print(x*100%3)

# power
print(x**10)

# abs
print((x-2).abs())
```

#### 2. Element wise operations

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print("a:\n", a)
print("b:\n", b)
```

```python
# add
print("a+b:\n", a+b)

# sub
print("a-b:\n", a-b)

# mul
print("a*b:\n", a*b)

# divide
print("a/b:\n", a/b)

# power
print("a**b:\n", a**b)

# mod
print("a%b:\n", a%b)
```

```python
# Absolute value
x = torch.randint(-100, 100, (2,3))
print(x)

abs_x = x.abs()
print(abs_x)
```

```python
# Negative
neg_x = torch.neg(x)
print(neg_x)
```

```python
# ceil
torch.ceil(x)
```

```python
# floor
torch.floor(x)
```

#### 4. Matrix Operations

```python
p = torch.randint(-10, 10, (3,4), dtype=torch.float32) + torch.rand(3, 4, dtype=torch.float32)
q = torch.randint(-10, 10, (4,3), dtype=torch.float32) + torch.rand(4, 3, dtype=torch.float32)

print("p:\n", p)
print("q:\n", q)
```

```python
# Matrix Multiplication
print("p@q:\n", torch.matmul(p, q))
print("p@q:\n", p@q)
```

```python
# Dot product
v1 = torch.tensor([1,2])
v2 = torch.tensor([4,5])

torch.dot(v1, v2)
```

```python
# Transpose
torch.transpose(p, 0, 1)
```

```python
# Determinant
torch.det(p@q)
```

```python
# Inverse
torch.inverse(p@q)
```

```python
# Test Inverse
torch.round(p@q@(torch.inverse(p@q)))
```

#### 5. Comparison Operations

```python
i, j = [torch.randint(-20, 20, (2,2)) for i in range(2)]
print("i:\n", i)
print("j:\n", j)
```

```python
# greater than
print(i > j)

# less than
print(i < j)

# Equal to
print(i == j)

# not equal to
print(i != j)

# greater than equal to
print(i >= j)

# less than equal to
print(i <= j)
```

### Tensors on GPU

```python
torch.cuda.is_available()
```

```python
device = torch.device("cuda")
```

```python
# Creating Tensor on GPU
x = torch.rand(2,3, device=device)
print(x)
```

```python
# CPU Tensor
a = torch.randint(-100, 100, (2,4))
print(a)
```

```python
# Moving an existing tensor to GPU
b = a.to(device=device)
print(b)
```

```python
# GPU Operation
b + 5
```

```python
import time
```

```python
# CPU vs GPU
size = 10000
mc1, mc2 = [torch.randn(size, size) for i in range(2)]
start_time = time.time()
mc = mc1@mc2
cpu_time = time.time() - start_time
```

```python
size = 10000
mg1, mg2 = [torch.randn(size, size, device=device) for i in range(2)]
start_time = time.time()
mg = mg1@mg2
gpu_time = time.time() - start_time
```

```python
# GPU is way faster.
cpu_time/gpu_time
```

### PyTorch to Numpy & Vice-Versa

```python
# Tensor to Numpy Array
import numpy as np

a = torch.randint(-100, 100, (4,4), dtype=torch.float32)
print(a)
print(type(a))

print("\n")

b = a.numpy()
print(b)
print(type(b))
```

```python
# Numpy Array to Tensor

c = torch.from_numpy(b)
print(c)
print(type(c))
```

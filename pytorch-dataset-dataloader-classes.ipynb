## Pytorch Dataset & Dataloader Classes

Until now we are loading the entire dataset to train the model, this could ne inefficient if the size of data is huge as memory is limited. Also, to get better and faster convergence the batches of data work well instead of providing the entire dataset at once, which is slow and inefficient.

Dataset and DataLoader are core abstractions in PyTorch that decouple how you define your data and how you efficiently iterate over it in the training loops.

```python
# Load Libraries

import torch
from torch.utils.data import Dataset, DataLoader
```

```python
# Data -> Dataset Class -> Dataloader Class
# Dataset Class - Loading data efficiently from the disk.
# DataLoader - Creating Batches to feed to training loop.
```

```python
# Dataset Class contains 3 methods:
# 1. Constructor (__init__()) - Which tells how data should be loaded.
# 2. Length (__len__()) - Which returns the total number of samples.
# 3. GetItem (__getitem__()) - Which returns data and label at the given index.
```

```python
# Dataloader Class
# Shuffle -> Chunk -> Batches -> Dataset GetItem -> Collate
```

```python
from sklearn.datasets import make_classification
import torch
```

```python
# Create a Synthetic Classification Dataset
X, y = make_classification(
    n_samples = 10,
    n_features = 2,
    n_informative = 2,
    n_redundant = 0,
    n_classes = 2,
    random_state = 42
)
```

```python
X = torch.from_numpy(X).to(dtype=torch.float32)
y = torch.from_numpy(y).to(dtype=torch.float32)
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
# Dataset Class
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels
    
    def __len__(self):
        return self.features.shape[0]

    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
len(dataset)
```

```python
# DataLoader Class

dataloader = DataLoader(dataset, batch_size = 2, shuffle=True)
```

```python
for batch_features, batch_labels in dataloader:
    print(batch_features)
    print(batch_labels)
    print("-"*50)
```

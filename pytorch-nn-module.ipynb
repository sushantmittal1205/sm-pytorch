### PyTorch nn Module

In last notebook, we created the Neural Network training pipeline and used raw code, now we'll use some PyTorch functionality to improve that.

We'll make following improvements:
1. Building Neural Network using nn module.
2. Use built-in activation function.
3. Use built-in loss function.
4. Use built-in optimizer

torch.nn Module (Build Neural Networks efficiently. It'll provide use pre-built layers, activation functions and loss functions.)

```python
import torch
import torch.nn as nn
import pandas as pd
```

```python
class SimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, features):
        out = self.linear(features)
        out = self.sigmoid(out)
        return out
```

```python
# Create Dataset
features = torch.rand(10, 5, dtype=torch.float32)

# Create Model
model = SimpleNN(features.shape[1])

# Forward Pass
model(features)
```

```python
# Using torchinfo to see Neural Network Summary
from torchinfo import summary
summary(model, input_size=(10, 5))
```

```python
# Lets create a Neural Network with Hidden Layer
class HLNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear1 = nn.Linear(num_features, 3)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(3, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, features):
        out = self.linear1(features)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)
        return out
```

```python
# Create Dataset
features = torch.rand(10, 5)

# Create Model
model = HLNN(features.shape[1])

# Forward Pass
model(features)
```

```python
# Model Summary
summary(model, input_size=(10, 5))
```

```python
# Using Sequential Container
class SeqHLNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features, 3),
            nn.ReLU(),
            nn.Linear(3, 1),
            nn.Sigmoid()
        )
    
    def forward(self, features):
        return self.network(features)
```

```python
# Create Dataset
features = torch.rand(10, 5)

# Create Model
model = SeqHLNN(features.shape[1])

# Forward Pass
model(features)
```

```python
# Model Summary
summary(model, input_size=(10, 5))
```

```python
# Now using what we have learned, we apply it to same dataset which we used in previous notebook.

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True)
df.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(df.drop(["diagnosis"], axis=1), df["diagnosis"], test_size=0.2, random_state=42)
# Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
# Move from Numpy Arrays to PyTorch Tensors
if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")

X_train_tensor = torch.from_numpy(X_train).to(device=device, dtype=torch.float32)
X_test_tensor = torch.from_numpy(X_test).to(device=device, dtype=torch.float32)
y_train_tensor = torch.from_numpy(y_train).to(device=device, dtype=torch.float32)
y_test_tensor = torch.from_numpy(y_test).to(device=device, dtype=torch.float32)
```

```python
# Model Class

class Model(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, X):
        out = self.linear(X)
        out = self.sigmoid(out)
        return out

    def loss_function(self, y_pred, y):
        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
        
        loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
        return loss 
```

```python
# Create Model

learning_rate = 0.1
epochs = 25
model = Model(X_train_tensor.shape[1])
model = model.to(device=device)

# Define Loop
for epoch in range(epochs):
    # Forward Pass
    y_pred = model(X_train_tensor)

    # Compute Loss
    loss = model.loss_function(y_pred, y_train_tensor)

    # Backward Pass
    loss.backward()

    # Update Parameters
    with torch.no_grad():
        model.linear.weight -= learning_rate * model.linear.weight.grad
        model.linear.bias -= learning_rate * model.linear.bias.grad

    # Zero Gradients
    model.linear.weight.grad.zero_()
    model.linear.bias.grad.zero_()

    # Print Loss Function
    print(f"Epoch: {epoch}, Loss: {loss}")
```

```python
# Evaluation
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f"Accuracy: {accuracy}")
```

```python
# We improved 2 things till now out of four:
# We are now using builtin nn Module
# We are now using builtin Activation Function
# Now two improvements are remaining - Built-in Activation Function and Built-in Optimizer.
```

```python
# Modified Model Class

class Model(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, X):
        out = self.linear(X)
        out = self.sigmoid(out)
        return out

loss_function = nn.BCELoss()
```

```python
# Create Model

learning_rate = 0.1
epochs = 25
model = Model(X_train_tensor.shape[1])
model = model.to(device=device)

# Define Loop
for epoch in range(epochs):
    # Forward Pass
    y_pred = model(X_train_tensor)

    # Compute Loss
    loss = loss_function(y_pred, y_train_tensor.view(-1, 1))

    # Backward Pass
    loss.backward()

    # Update Parameters
    with torch.no_grad():
        model.linear.weight -= learning_rate * model.linear.weight.grad
        model.linear.bias -= learning_rate * model.linear.bias.grad

    # Zero Gradients
    model.linear.weight.grad.zero_()
    model.linear.bias.grad.zero_()

    # Print Loss Function
    print(f"Epoch: {epoch}, Loss: {loss}")
```

```python
# Evaluation
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f"Accuracy: {accuracy}")
```

```python
# Now we improved by adding Built-in Loss function
# Last modification remaining is to use built-in optimizer (We are going to use torch.optim for that.)

# torch.optim is module in PyTorch that provides variety of optimization algorithms.
# Some common are Stochastic GD, Adam, RMSprop
```

```python
# Final Model (All 4 improvements)

class FModel(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(num_features, 1),
            nn.Sigmoid()
        )
    
    def forward(self, X):
        return self.layers(X)

model = FModel(X_train_tensor.shape[1])
model = model.to(device=device)

learning_rate = 0.1
epochs = 25

# Define Loss Function
loss_function = nn.BCELoss()

# Define Optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# Define Loop
for epoch in range(epochs):
    # Forward Pass
    y_pred = model(X_train_tensor)

    # Compute Loss
    loss = loss_function(y_pred, y_train_tensor.view(-1, 1))

    # Zero Gradients
    optimizer.zero_grad()

    # Backward Pass
    loss.backward()

    # Update Parameters
    optimizer.step()

    # Print Loss Function
    print(f"Epoch: {epoch}, Loss: {loss}")
```

```python
# Evaluation
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f"Accuracy: {accuracy}")
```
